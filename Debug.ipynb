```python
def format_params(number, precision=2, exponent_digits=2):
    # Format the number in scientific notation
    scientific = f"{number:.{precision}e}"
    # Split into base and exponent
    base, exponent = scientific.split("e")
    # Reformat the exponent with leading zeros
    exponent = int(exponent)  # Convert to integer
    formatted_exponent = f"e{'+' if exponent >= 0 else '-'}{abs(exponent):0{exponent_digits}d}"
    return f"{base}{formatted_exponent}"

#format_params(temperature, 0, 1)
```

```python
from scipy.fftpack import fftn
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import yt

############    Basic system params, modify if needed    ############################
step1 = 10000
step2 = 20000
plot_int = 10
nsteps = (step2-step1)//plot_int + 1
temperature = 1e-5
alpha0 = 0.85
T = 1e-5
cs2 = 1./3.
#####################################################################################

################################    Load Data   ########################################
plot_file_dir = "./data_mixture/lbm_data_shshan_alpha0_" + "{:.2f}".format(alpha0) +\
          "_xi_" + "{:.1e}".format(temperature) + "_size64-4-4_continue/plt" + f"{step1:06d}"
ds = yt.load(plot_file_dir)
# extract multifab component specified by [rhoA] and convert it to numpy ndarray
profile_data = ds.all_data()[("boxlib", "rho")].to_ndarray()
domain_size = ds.domain_dimensions
#print(profile_data.reshape(domain_size))

rho_frames = np.zeros((domain_size[0],domain_size[1],domain_size[2],nsteps))
rho_frames_mean = np.zeros(domain_size)
for n in range(step1, step2+plot_int, plot_int):
    plot_file_dir = "./data_mixture/lbm_data_shshan_alpha0_" + "{:.2f}".format(alpha0) +\
          "_xi_" + "{:.1e}".format(temperature) + "_size64-4-4_continue/plt" + f"{n:06d}" 
    ds = yt.load(plot_file_dir)
    # extract multifab component specified by [rhoA] and convert it to numpy ndarray
    profile_data_flat = ds.all_data()[("boxlib", "rho")].to_ndarray()
    rho_frames[:,:,:,(n-step1)//plot_int] = profile_data_flat.reshape(domain_size)
rho_frames_mean = np.mean(rho_frames, 3)
```

```python
N = domain_size[0]*domain_size[1]*domain_size[2]
rho_frames_variation = np.zeros(((domain_size[0],domain_size[1],domain_size[2],nsteps)))
for n in range(step1, step2+plot_int, plot_int):
    # sustracting mean values from each frame
    rho_frames_variation[:,:,:,(n-step1)//plot_int] = rho_frames[:,:,:,(n-step1)//plot_int] - rho_frames_mean
print("mean density rho variation is {}\n".format(np.mean(rho_frames_variation)))

delta_rho_frames_dft = np.zeros((domain_size[0], domain_size[1], domain_size[2], nsteps), dtype=complex)
for n in range(step1, step2+plot_int, plot_int):
    # DFT for each frame
    delta_rho_frames_dft[:,:,:,(n-step1)//plot_int] = fftn(rho_frames_variation[:,:,:,(n-step1)//plot_int])
delta_rho_frames_dft = delta_rho_frames_dft/np.sqrt(N)    # normalization

delta_rho_frames_dft_norm2 = np.zeros(domain_size)
for n in range(0, nsteps):
    norm = np.abs(delta_rho_frames_dft[:,:,:,n])
    delta_rho_frames_dft_norm2 = delta_rho_frames_dft_norm2 + norm*norm
delta_rho_frames_dft_norm2 = delta_rho_frames_dft_norm2/nsteps/(T/cs2)

for i in range(0, domain_size[0]):
    for j in range(0, domain_size[1]):
        for k in range(0, domain_size[2]):
            print("({},{},{})={}".format(i, j, k, delta_rho_frames_dft_norm2[i,j,k]))

```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import os
from PIL import Image
def read_images_from_directory(directory):
    images = []
    for filename in os.listdir(directory):
        if filename.endswith(".jpg") or filename.endswith(".png"):  # Check for image file extensions
            filepath = os.path.join(directory, filename)
            images.append(filepath)
    return images

fig, ax = plt.subplots()

#img = plt.imread("./tau_2_alpha_3_xi_0/48.png")

images = []
images_path = read_images_from_directory("./IMG_shshan_alpha_0_xi_0/")
for filePath in images_path:
    img = plt.imread(str(filePath))
    im = ax.imshow(img, animated=True)
    images.append([im])

ani = animation.ArtistAnimation(fig, images, interval=500, blit=True, repeat_delay=1000)
writergif = animation.PillowWriter(fps=3)
ani.save('filename.gif',writer=writergif)
plt.show()
```

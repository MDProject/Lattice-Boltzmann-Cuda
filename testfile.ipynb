```python
#########################################   DFT     ##################################################
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft2, fftshift, fftn

# Create a 2D sine-cosine signal
N = 8  # Grid size
x = np.linspace(0, N-1, N)
y = np.linspace(0, N-1, N)
X, Y = np.meshgrid(x, y)
print(x)
# Define a composite sine-cosine wave
signal = np.sin(2 * np.pi/N* X)# + np.cos(2 * np.pi/N * Y)

# Compute the 2D FFT
fft2_result = fftn(signal)

print(fft2_result)
fft2_shifted = fftshift(fft2_result)  # Shift the zero-frequency to the center
magnitude_spectrum = np.log(np.abs(fft2_shifted) + 1)  # Log scale for better visibility

# Plot the original signal and its Fourier Transform
plt.figure(figsize=(12, 6))

# Original signal
plt.subplot(1, 2, 1)
plt.imshow(signal, extent=(0, 2 * np.pi, 0, 2 * np.pi), cmap='viridis', origin='lower')
plt.title("Original Signal (Sine + Cosine)")
plt.xlabel("X")
plt.ylabel("Y")

# Magnitude Spectrum
plt.subplot(1, 2, 2)
plt.imshow(magnitude_spectrum, cmap='gray', extent=(-N//2, N//2, -N//2, N//2), origin='lower')
plt.title("2D Fourier Transform Magnitude Spectrum")
plt.xlabel("Frequency X")
plt.ylabel("Frequency Y")

plt.tight_layout()
plt.show()
```
